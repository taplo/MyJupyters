#### 获得各个时期的分布函数

```python
env=sp.stats.gaussian_kde(dp.close)
env01=sp.stats.gaussian_kde(dp.close[:30])
env02=sp.stats.gaussian_kde(dp.close[30:60])
env03=sp.stats.gaussian_kde(dp.close[60:90])
env04=sp.stats.gaussian_kde(dp.close[90:120])
env05=sp.stats.gaussian_kde(dp.close[120:150])
env06=sp.stats.gaussian_kde(dp.close[150:180])
env07=sp.stats.gaussian_kde(dp.close[180:210])
env08=sp.stats.gaussian_kde(dp.close[210:240])
env09=sp.stats.gaussian_kde(dp.close[240:270])
env10=sp.stats.gaussian_kde(dp.close[270:300])
```

#### 建立DataFrame

```python
dex=np.arange(600,6000,1)
list=pd.DataFrame(index=dex)
list['org']=env.evaluate(list.index)
list['01']=env01.evaluate(list.index)
list['02']=env02.evaluate(list.index)
list['03']=env03.evaluate(list.index)
list['04']=env04.evaluate(list.index)
list['05']=env05.evaluate(list.index)
list['06']=env06.evaluate(list.index)
list['07']=env07.evaluate(list.index)
list['08']=env08.evaluate(list.index)
list['09']=env09.evaluate(list.index)
list['10']=env10.evaluate(list.index)
```

```python
list[:5]
```

#### 显示图形

```python
list.plot(grid=True,figsize=(16,9),xticks=cxt,xlim=cxl)
```
