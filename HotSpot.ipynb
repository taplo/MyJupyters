导入模块

```python
import tushare as ts
import datetime as dt
```

## 积极信息

### 高转送股票信息

```python
df = ts.profit_data()
if len(df)<2:
	print "无！"
else:
	df=df.sort_values('shares',ascending=False)
	print df[:20]
```

### 业绩预披露靠前

```python
year=dt.date.today().year
month=dt.date.today().month
if month>9:
    season=4
elif month>6:
    season=3
elif month>3:
    season=2
else:
    season=1

if season==1:
    year=year-1
    season=4
else:
    season=season-1

dp=ts.forecast_data(year,season)
dp=dp.sort_values('range',ascending=False)
print u"\n本期预披露靠前的结果："
print '\n',dp[:30]

#上期预披露结果
if season==1:
    year=year-1
    season=4
else:
    season=season-1

dp=ts.forecast_data(year,season)
dp=dp.sort_values('range',ascending=False)
print u"\n上期预披露靠前的结果："
print dp[:20]
```

### 每日龙虎榜列表

```python
day=dt.date.today()
while ts.is_holiday(str(day)):
	day=day-dt.timedelta(days=1)
df=ts.top_list(str(day))
print df
```

### 个股上榜统计

```python
df=ts.cap_tops()
df=df.sort_values('count',ascending=False)
print '\n',df[:20]
```

### 机构席位追踪

```python
df=ts.inst_tops()
df=df.sort_values('bcount',ascending=False)
print '\n',df[:20]
```

### 机构成交明细

```python
df=ts.inst_detail()
df=df.sort_values('date',ascending=False)
print '\n',df[:50]
```

```python
df=ts.xsg_data()
df['ratio']=df['ratio'].astype(float)
df=df.sort_values('ratio',ascending=False)
print df[df.ratio>10]
```
